# Scaling - unskaling pipeline for manual multipliers

```python
mults = {
    'dens': 1e-5,
    'zip_area': 5e2,
    'zip_long': .5,
    'long_mean': .5,
    'number_of_panels_north': 3e-3,
    'number_of_panels_south': 5e-4,
    'number_of_panels_east': 1e-3,
    'number_of_panels_west': 7e-4,
    'number_of_panels_flat': 3e-4,
    'number_of_panels_total': 1e-4,
    'potential_kwh_total': 5e-7,
    'long_min': .5,
    'long_max': .5,
    'state_long': .5,
    'state_size': 10,
    'is_usa': 50
}

class ManualScaler:
    def __init__(self, mults, label_col: str):
        self.mults = mults
        self.label_col = label_col
        
    def X_scale(self, df: pd.DataFrame):
        """Function applies multipliers from the passed dict to respective columns of the passed df"""
        local_df = df.copy()
        for k in self.mults:
            local_df.loc[:, k] = local_df[k] * mults[k]
        return local_df
    
    def X_unscale(self, df: pd.DataFrame):
        """Function unscales the passed df using self.mults"""
        local_df = df.copy()
        for k in self.mults:
            local_df.loc[:, k] = local_df[k] / mults[k]
        return local_df
    
    def y_unscale(self, y_scaled: np.array):
        return y_scaled / self.mults[self.label_col]
    
scaler = ManualScaler(mults=mults, label_col='potential_kwh_total')
```

# Lasso vs Ridge on manually scaled dataset

```python
sns.boxplot(scaled_df)
plt.xticks(rotation = 85)
plt.ylim(bottom=-150, top=250)
```

```python
mean_res = []
lasso_weights = []
for _ in range(100):
    X_train, X_test, y_train, y_test = train_test_split(scaled_df[[col for col in list(scaled_df) if col!=scaled_label]].values, 
                                                        scaled_df[[scaled_label]].values, test_size=.2)
    lasso_model = Lasso()#linear_model.LinearRegression(positive=True)
    lasso_model.fit(X_train, y_train)
    y_pred = lasso_model.predict(X_test)
    mean_res.append(np.sqrt(mean_squared_error(y_pred=y_pred, y_true=y_test)) / np.std(y_test))
    lasso_weights.append(lasso_model.coef_)
np.mean(mean_res)


```

```python
mean_res = []; ridge_weights = []
for _ in range(50):
    X_train, X_test, y_train, y_test = train_test_split(scaled_df[[col for col in list(scaled_df) if col!=scaled_label]].values, 
                                                        scaled_df[[scaled_label]].values, test_size=.2)
    ridge_model = Ridge()#linear_model.LinearRegression(positive=True)
    ridge_model.fit(X_train, y_train)
    y_pred = ridge_model.predict(X_test)
    mean_res.append(np.sqrt(mean_squared_error(y_pred=y_pred, y_true=y_test)) / np.std(y_test))
    ridge_weights.append(ridge_model.coef_)
np.mean(mean_res)


```

# Simple linear model on robust scaled dataset

```python
def norm_rmse(estimator, X_test, y_test):
    y_pred = estimator.predict(X_test)
    naive_baseline_rmse = np.sqrt(mean_squared_error(y_pred=y_pred, y_true=y_test))
    return {'nrmse': naive_baseline_rmse / np.std(y_test),
            'r2': r2_score(y_pred=y_pred, y_true=y_test)}
```

```python
kfold = RepeatedKFold(n_splits=5, n_repeats=100)
test_res = cross_validate(Lasso(), 
                          X=small_df[[col for col in list(small_df) if col!=small_label]].values, 
                          y=small_df[[small_label]].values,
                          cv=kfold,
                          scoring = norm_rmse)
print(f"NRMSE: {test_res['test_nrmse'].mean()}\n R2: {test_res['test_r2'].mean()}") #
```
